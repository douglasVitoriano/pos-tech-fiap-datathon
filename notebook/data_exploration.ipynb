### Pre-processamento

Como JSON não é apropriado para treinamento de modelos, é realizado uma conversao para csv

```python
data = []

for key, value in df_applicants.iterrows():
    record = {
        "codigo_profissional": value["infos_basicas"]["codigo_profissional"],
        "nome": value["infos_basicas"]["nome"],
        "telefone": value["infos_basicas"]["telefone"],
        "email": value["infos_basicas"]["email"],
        "objetivo_profissional": value["infos_basicas"]["objetivo_profissional"],
        "data_criacao": value["infos_basicas"]["data_criacao"],
        "data_atualizacao": value["infos_basicas"]["data_atualizacao"],
        "telefone_celular": value["informacoes_pessoais"]["telefone_celular"],
        "data_nascimento": value["informacoes_pessoais"]["data_nascimento"],
        "formacao_academica": value["formacao_e_idiomas"]["nivel_academico"],
        "idioma_ingles": value["formacao_e_idiomas"]["nivel_ingles"],
        "idioma_espanhol": value["formacao_e_idiomas"]["nivel_espanhol"],
        "cv_pt": value["cv_pt"]
    }
    data.append(record)


df_applicants = pd.DataFrame(data)
```

```python
df_applicants.head()
```

```python
df_applicants.to_csv("../dados/trusted/applicants_trusted.csv", index=False)
```

## Prospects

```python
df_prospects = pd.read_json("../dados/raw/prospects.json").T
```

```python
df_prospects.head()
```

```python
data = []

for index, row in df_prospects.iterrows():
    for prospect in row["prospects"]:
        record = {
            "titulo": row["titulo"],
            "nome": prospect["nome"],
            "codigo": prospect["codigo"],
            "situacao_candidado": prospect["situacao_candidado"],
            "data_candidatura": prospect["data_candidatura"],
            "ultima_atualizacao": prospect["ultima_atualizacao"],
            "comentario": prospect["comentario"],
            "recrutador": prospect["recrutador"]
        }
        data.append(record)
    data.append(record)


df_prospects = pd.DataFrame(data)
```

```python
df_prospects.head()
```

```python
df_prospects.to_csv("../dados/trusted/prospects_trusted.csv", index=False)
```
